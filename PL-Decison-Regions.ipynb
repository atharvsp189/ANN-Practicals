```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(
    n_samples=80, # No of data points
    n_features=2, 
    n_classes=2, 
    n_clusters_per_class=1, 
    n_redundant=0, 
    random_state=42
)
```

```python
weights = np.random.rand(2)
bias = np.random.rand()
learning_rate = 0.01

print("Initial Weights:", weights)
print("Initial Bias:", bias)

def activation_function(x):
    return 1 if x>=0 else 0
```

```python
errors = []
epochs=100

for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        net_input = np.dot(X[i], weights) + bias
        output = activation_function(net_input)
        error = Y[i] - output
        total_error += abs(error)
        weights += learning_rate * error * X[i]
        bias += learning_rate * error
        errors.append(total_error)
```

```python
plt.figure(figsize=(8, 6))

# Scatter plot for each class without loop
plt.scatter(X[Y==0, 0], X[Y==0, 1], color='red', marker='o', label='Class 0')
plt.scatter(X[Y==1, 0], X[Y==1, 1], color='blue', marker='x', label='Class 1')

# Decision boundary
x_values = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)
y_values = -(weights[0] * x_values + bias) / weights[1]
plt.plot(x_values, y_values, 'g--', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```
